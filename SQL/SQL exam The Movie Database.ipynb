## Question 1
Who won the Oscar for “Actor in a Leading Role” in 2015?

(Hint: The winner is indicated as '1.0'.)

Options:

Micheal Fassbender
Natalie Portman
Leonardo DiCaprio
Eddie Redmayne

```python
%%sql

SELECT  name
FROM Oscars
WHERE year = '2015'
  AND award = 'Actor in a Leading Role'
  AND winner = '1.0';
```

```python
%%sql
SELECT COUNT(DISTINCT award) AS unique_awards
FROM Oscars;
```

```python
%%sql
SELECT COUNT(movie_id) AS movie_count
FROM Movies
WHERE title LIKE '%Spider%';
```

```python
%%sql
SELECT COUNT(DISTINCT T1.movie_id) AS movie_count
FROM Movies AS T1
INNER JOIN Genremap AS T2 ON T1.movie_id = T2.movie_id
INNER JOIN Genres AS T3 ON T2.genre_id = T3.genre_id
INNER JOIN Keywordmap AS T4 ON T1.movie_id = T4.movie_id
INNER JOIN Keywords AS T5 ON T4.keyword_id = T5.keyword_id
WHERE T3.genre_name = 'Thriller'
  AND T5.keyword_name LIKE '%love%';
```

```python
%%sql
SELECT COUNT(movie_id) AS movie_count
FROM Movies
WHERE release_date BETWEEN '2006-08-01' AND '2009-10-01'
  AND popularity > 40
  AND budget < 50000000;
```

```python
%%sql
SELECT COUNT(DISTINCT T2.characters) AS unique_characters
FROM Actors AS T1
INNER JOIN Casts AS T2 ON T1.actor_id = T2.actor_id
WHERE T1.actor_name = 'Vin Diesel';
```

```python
%%sql
SELECT T3.genre_name
FROM Movies AS T1
INNER JOIN Genremap AS T2 ON T1.movie_id = T2.movie_id
INNER JOIN Genres AS T3 ON T2.genre_id = T3.genre_id
WHERE T1.title = 'The Royal Tenenbaums';
```

```python
%%sql
SELECT
    T3.production_company_name,
    AVG(T1.popularity) AS avg_popularity
FROM
    Movies AS T1
INNER JOIN
    productioncompanymap AS T2 ON T1.movie_id = T2.movie_id
INNER JOIN
    productioncompanies AS T3 ON T2.production_company_id = T3.production_company_id
GROUP BY
    T3.production_company_name
ORDER BY
    avg_popularity DESC
LIMIT 3;
```

```python
%%sql
SELECT COUNT(actor_id) AS actor_count
FROM Actors
WHERE gender = 1
  AND actor_name LIKE 'N%';
```

```python
%%sql
SELECT T3.genre_name
FROM Movies AS T1
INNER JOIN Genremap AS T2 ON T1.movie_id = T2.movie_id
INNER JOIN Genres AS T3 ON T2.genre_id = T3.genre_id
GROUP BY T3.genre_name
ORDER BY AVG(T1.popularity) ASC
LIMIT 1;
```
